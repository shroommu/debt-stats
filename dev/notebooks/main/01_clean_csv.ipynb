# Clean CSV Data

### Common Imports

```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog
import cloudpickle as cpkl
```

### Load Data

```python
csv_file_path = filedialog.askopenfilename(
    title="Select CSV file", filetypes=[("CSV files", "*.csv")]
)

raw_df = pd.read_csv(csv_file_path)
```

### Dataframe Info

```python
raw_df.info()
```

```python
raw_df.describe()
```

```python
df = raw_df.copy()
```

```python
df.columns = df.columns.str.strip().str.replace("Q4_", "")
```

```python
num_cols = df.columns[df.columns != "state"]

df[num_cols] = (
    df[num_cols]
    .replace(",", "", regex=True)
    .fillna("0")
    .apply(pd.to_numeric, errors="coerce")
    .astype(int)
)
```

```python
df["state"] = df["state"].str.strip()
```

```python
save_path = filedialog.askdirectory(title="Save cleaned CSV file")

file_name = csv_file_path.split("/")[-1].replace(".csv", "_cleaned.csv")

df.to_csv(f"{save_path}/{file_name}", index=False)
```
